```python
from numpy import *
from scipy.optimize import curve_fit
from matplotlib.pyplot import *
```

```python
def linear(t,a,b):
  return (a*t+b)

def quadratic(t,a,b,c):
  return (a*pow(t,2) + b*t +c)

def cubic(t,a,b,c,d):
  return (a*pow(t,3) + b*pow(t,2) + c*t +d)

def quartic(t,a,b,c,d,e):
  return (a*pow(t,4) + b*pow(t,3) + c*pow(t,2) + d*t +e)
```

```python
def read_file():
  temprature = []
  cp = []
  for line in open('data','r'):
    values = line.split(',')
    temprature.append(float(values[1]))
    cp.append(float(values[2]))
  return[temprature,cp]
```

```python
temprature,cp = read_file()
popt,pcov = curve_fit(linear,temprature,cp)
fit_cp1 = linear(array(temprature), *popt)

temprature,cp = read_file()
popt,pcov = curve_fit(quadratic,temprature,cp)
fit_cp2 = quadratic(array(temprature), *popt)

temprature,cp = read_file()
popt,pcov = curve_fit(cubic,temprature,cp)
fit_cp3 = cubic(array(temprature), *popt)

temprature,cp = read_file()
popt,pcov = curve_fit(quartic,temprature,cp)
fit_cp4 = quartic(array(temprature), *popt)
```

```python
plot(temprature,cp,label="Actual Data")
plot(temprature,fit_cp1,label="Linear Curve Fit")
legend()
title("Linear Curve Fit")
xlabel("temprature (in k)")
ylabel("Specific heat(in KJ/kg-K")
savefig("Linear_Curve_Fit.jpg")
show()

plot(temprature,cp,label="Actual Data")
plot(temprature,fit_cp2,label="Quadratic Curve Fit")
legend()
title("Quadratic Curve Fit")
xlabel("temprature (in k)")
ylabel("Specific heat(in KJ/kg-K")
savefig("Quadratic_Curve_Fit.jpg")
show()

plot(temprature,cp,label="Actual Data")
plot(temprature,fit_cp3,label="Cubic Curve Fit")
legend()
title("Cubic Curve Fit")
xlabel("temprature (in k)")
ylabel("Specific heat(in KJ/kg-K")
savefig("Cubic_Curve_Fit.jpg")
show()

plot(temprature,cp,label="Actual Data")
plot(temprature,fit_cp4,label="Quartic Curve Fit")
legend()
title("Quartic Curve Fit")
xlabel("temprature (in k)")
ylabel("Specific heat(in KJ/kg-K")
savefig("Quartic_Curve_Fit.jpg")
show()

```
